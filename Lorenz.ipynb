```python
from scipy.integrate import solve_ivp
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Constants
sigma = 18
rho   = 28
beta  = 2.4
t_max = 50
```

```python
def lorenz(t, xyz):
    x, y, z = xyz
    dx = sigma * (y - x)
    dy = x * (rho - z) - y
    dz = x * y - beta * z
    return dx, dy, dz
```

```python
xyz_init = (-10, -10, 35)
t_span = (0., t_max)
t_values = np.arange(0, t_max, 0.001)
```

```python
solution = solve_ivp(lorenz, t_span, xyz_init, t_eval=t_values)
xyz = solution.y
```

```python
x, y, z = xyz
```

```python
plt.figure(figsize=(12,8))
ax = plt.axes(projection='3d')
ax.plot3D(x, y, z)
plt.show()
```
