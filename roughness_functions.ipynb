# Stabilized roughness functions

### Imports:

```python
from scipy import special
import numpy as np
import matplotlib.pyplot as plt
```

### Input array:

```python
x = np.linspace(-3, 3)
```

### Definition of the stabilized roughness functions:

```python
%%latex
\begin{align}
h_+(x) &= \frac{1}{2} \exp\left(\frac{x^2}{2}\right) erfc\left(-\frac{x}{\sqrt{2}}\right) \\
h_-(x) &= \frac{1}{2} \exp\left(\frac{x^2}{2}\right) erfc\left(+\frac{x}{\sqrt{2}}\right) \\
H(x) &= \exp\left(\frac{x^2}{2}\right)
\end{align}
```

```python
def hplus(x_in):
    return np.exp(x_in**2/2)*special.erfc(-x_in/np.sqrt(2))/2
```

```python
def hmin(x_in):
    return np.exp(x_in**2/2)*special.erfc(+x_in/np.sqrt(2))/2
```

### Plot both functions together with unstabilized version:

```python
plt.figure(figsize=(12, 8))
plt.plot(x, hplus(x), label='h+')
plt.plot(x, hmin(x), label='h-')
plt.plot(x, np.exp(x**2/2), label='H')
plt.legend()
plt.ylim((0,5))
plt.show()
```

### Same plots, but on a logarithmic scale:

```python
plt.figure(figsize=(12, 8))
plt.semilogy(x, hplus(x), label='h+')
plt.semilogy(x, hmin(x), label='h-')
plt.semilogy(x, np.exp(x**2/2), label='H')
plt.legend()
plt.show()
```
